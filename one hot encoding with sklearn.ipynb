### One word at a time

```python
import pandas as pd

text = ['cat', 'dog', 'fish', 'bird', 'cat', 'dog']

data = pd.get_dummies(text)

print(data)
print('-'*30)
print(data.keys())
```

### Whole sentence at a time

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

text = ['i love dogs', 'i dont mind cats', 'turtles are alright']

text = np.reshape(text,(-1,1))

encoder = OneHotEncoder()
encoder.fit(text)

one_hot = encoder.transform(text)

print(one_hot.toarray())
print([i[3:] for i in encoder.get_feature_names_out()])
```

### Tokenized sentence

```python
from sklearn.preprocessing import OneHotEncoder

text = [['this', 'is', 'bad'], ['but', 'its', 'fine']]

encoder = OneHotEncoder()
encoder.fit(text)

one_hot = encoder.transform(text)

print(one_hot.toarray())
print('-'*30)
print([i[3:] for i in encoder.get_feature_names_out()])
```

```python
pd.DataFrame(one_hot.toarray(), columns= [i[3:] for i in encoder.get_feature_names_out()])
```
