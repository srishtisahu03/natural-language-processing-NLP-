```python
from nltk.tokenize import word_tokenize, sent_tokenize

corpus = "GeeksforGeeks is not a website or a company. GeeksforGeeks is a coding environment. Thats what i love. Which is good. Not bad at all"
```

### Finding the sequence

```python
seq = set()
#set because a set can only store unique values whereas a list can repeat values

for word in word_tokenize(corpus):
    if (word != '.'):
        seq.add(word)
        
seq = list(seq)

print(seq)
```

```python
len(seq)
```

```python
seq.index('GeeksforGeeks')
```

### Finding index to encode

```python
data = []

for sent in sent_tokenize(corpus):
    index = []
    for word in word_tokenize(sent):
        if (word != '.'):
            index.append(seq.index(word))
            
    data.append(index)
```

```python
data;
```

```python
enc = [0 for x in range(len(seq))]
enc
```

### Encoding the sentences in one hot encoding format

```python
fin = []

for indexes in data:
    enc = [0 for x in range(len(seq))]
    for index in indexes:
        enc[index] = 1
    fin.append(enc)
```

```python
fin;
```

### Creating the dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame(fin, columns=seq)
```
