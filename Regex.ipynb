#### Findall

```python
a = "hiiii how are you area arena aRe"
b = "are"
```

```python
re.findall(b,a)
```

#### Finditer

```python
for s in re.finditer("are","hiiii how are you area arena aRe"):
    print(s.span())
```

#### Creating regular expressions

```python
txt = "My telephone number is 116"

pattern = "\d\d\d"

re.search(pattern, txt)
```

```python
txt = "My telephone number is 1-16"

pattern = "\d-\d\d"

re.search(pattern, txt)
```

```python
txt = "My telephone number is 116"

pattern = "\d{3}"

re.search(pattern, txt)
```

```python
txt = "My telephone number is 116"

pattern = "\w\w\w"

re.search(pattern, txt)
```

```python
re.search(pattern, txt).group()
```

#### Start and end pattern matching

```python
re.findall("\d","4 is divisible by 2 not 33")
```

```python
# last digit
re.findall("\d$","4 is divisible by 2 not 33")
```

```python
# First digit
re.findall("^\d","4 is divisible by 2 not 33")
```

#### Finding capital, small letters and numbers

```python
re.findall("[A-Z]", "Whats Up Dawg03")
```

```python
re.findall("[a-z]", "Whats Up Dawg03")
```

```python
re.findall("[A-Za-z]", "Whats Up Dawg03")
```

```python
re.findall("[A-Za-z0-9]", "Whats Up Dawg03")
```

#### Removing special characters

```python
txt = "Wha@t$s up! da#wg?"
```

```python
''.join(re.findall('[^@$!#?]+',txt))
```

#### Exclusion

```python
print(''.join(re.findall("\D","4 is divisible by 2 not 33")))
```

#### Phone number pattern matching

```python
for no in ["334-5436346-345","345-2357-6567","0-27886-4327928"]:
    print(str(re.findall('[\d]+-[\d]+-[\d]', no)[0]).replace('-',''))
```

#### Email pattern matching

```python
mail = "srishti@gmail.com"
p = "[A-Za-z0-9]+@[\w]+.[\w]+"
re.search(p,mail)
```

```python
re.search(p,mail).group()
```

#### Email id with specific domain

```python
mail = "srishti@gmail.com"
p = "[A-Za-z0-9]+@(gmail.com)$"
re.search(p,mail)
```

#### Continuous numbers

```python
txt = "the number is 9374392 not 39272"

c = 0

for i in range(len(txt)):
    if(txt[i].isdigit() == True):
        c += 1
        print(txt[i], c, i)
    else:
        c=0
```
